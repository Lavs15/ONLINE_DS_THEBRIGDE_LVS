```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from statsmodels.tsa.stattools import grangercausalitytests
```

```python
df_fiat = pd.read_csv('fiat.csv')
df_crypto = pd.read_csv('crypto.csv')
df_crypto_meme = pd.read_csv('crypto_meme.csv')
df_crypto_ia = pd.read_csv('crypto_ia.csv')
```

```python
df_fiat
```

```python
df_fiat['Date'] = pd.to_datetime(df_fiat['Date'],utc=True)
df_fiat['Date'] = df_fiat['Date'].dt.date
df_crypto['Date'] = pd.to_datetime(df_crypto['Date'],utc=True)
df_crypto['Date'] = df_crypto['Date'].dt.date
df_crypto_meme['Date'] = pd.to_datetime(df_crypto_meme['Date'],utc=True)
df_crypto_meme['Date'] = df_crypto_meme['Date'].dt.date
df_crypto_ia['Date'] = pd.to_datetime(df_crypto_ia['Date'],utc=True)
df_crypto_ia['Date'] = df_crypto_ia['Date'].dt.date
```

```python
df_fiat.set_index('Date', inplace=True)
df_crypto.set_index('Date', inplace=True)
df_crypto_meme.set_index('Date', inplace=True)
df_crypto_ia.set_index('Date', inplace=True)
```

```python
df_all = df_fiat
df_all = pd.merge(df_all, df_crypto, left_index=True, right_index=True)
df_all = pd.merge(df_all, df_crypto_meme, left_index=True, right_index=True)
df_all = pd.merge(df_all, df_crypto_ia, left_index=True, right_index=True)

df_all
```

```python
df_fiat1 = df_all[df_fiat.columns]
df_crypto1 = df_all[df_crypto.columns]
df_crypto_meme1 = df_all[df_crypto_meme.columns]
df_crypto_ia1 = df_all[df_crypto_ia.columns]
```

```python
def plot_data(df, title, y_label = "Precio en USD", y_limit=None):
    plt.figure(figsize=(12, 6))
    for column in df.columns:
        plt.plot(df.index, df[column], label=column)
    
    plt.title(title)
    plt.xlabel('Fecha')
    plt.ylabel(y_label)
    if y_limit:
        plt.ylim(y_limit)
    plt.legend()
    plt.grid(True)
    plt.savefig(f'{title}.png')
    plt.show()
   

plot_data(df_fiat1, 'Fiat Precio en el tiempo')
```

```python
df_fiat_norm = df_fiat1 / df_fiat1.iloc[0]
plot_data(df_fiat_norm, 'FIAT Precio normalizado en el tiempo', y_label='Precio normalizado')
```

```python
df_crypto_norm = df_crypto1 / df_crypto1.iloc[0]
plot_data(df_crypto_norm,'Cripto Precio normalizado en el tiempo', y_label='Precio normalizado')
```

```python
df_crypto_ia_norm = df_crypto_ia1 / df_crypto_ia1.iloc[0]
plot_data(df_crypto_ia_norm, 'Cripto IA Precio normalizado en el tiempo', y_label='Precio normalizado')
```

```python
df_crypto_meme_norm = df_crypto_meme1 / df_crypto_meme1.iloc[0]
plot_data(df_crypto_meme_norm, 'Cripto MEME Precio normalizado en el tiempo', y_label='Precio normalizado')
```

```python
df_fiat_log = np.log(df_fiat1)
df_fiat_log1 = df_fiat_log - df_fiat_log.shift(1)
df_fiat_log1 = df_fiat_log1.dropna()

plot_data(df_fiat_log1, 'Log Fiat Price Over Time', y_label='Log Price')
plot_data(df_fiat_log, 'Normalized Fiat Price Over Time', y_label='Normalized Price')
```

```python
max_lag = 5
p_value = {}
for target in df_fiat_log1.columns:
    test_result = grangercausalitytests(df_fiat_norm[[target, 'Euro']], maxlag=max_lag, verbose=False)
    pvals = [test_result[lag][0]['ssr_ftest'][1] for lag in range(1, max_lag + 1)]
    p_value[target] = min(pvals) 
p_value_df = pd.DataFrame(p_value, index=[0]).T
p_value_df
```

```python
from scipy.stats import pearsonr
matriz_pvalor = {}
for i in df_fiat1.columns:
    for j in df_fiat1.columns:
        if i != j:
            corr, p_value = pearsonr(df_fiat1[i], df_fiat1[j])
            matriz_pvalor[(i, j)] = (corr, p_value)
    
matriz_pvalor_df = pd.DataFrame(matriz_pvalor, index=['Correlation', 'P-Value']).T
matriz_pvalor_df
```
